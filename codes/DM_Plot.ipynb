```python
import pandas as pd
```

```python
df=pd.read_csv("All data - Average_error.csv")
```

```python
df_cp=pd.read_csv("DM_DATA.csv")
```

```python
df_cp = df_cp.dropna()
df_cp
```

```python
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

# Create scatter plot with regression line
sns.regplot(x='PREDICTION ERROR', y='O-LIFE SCORE', data=df_cp)
x= df_cp['PREDICTION ERROR']
y=df_cp['O-LIFE SCORE']

# Calculate correlation coefficient and p-value
correlation_coefficient = df_cp['PREDICTION ERROR'].corr(df_cp['O-LIFE SCORE'])
p_value = stats.pearsonr(x, y)[1]

# Annotate the plot with correlation coefficient and p-value
plt.text(0.5, 0.9, f'Correlation Coefficient: {correlation_coefficient:.2f}\nP-value: {p_value:.2f}',
         horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)

# Add labels to the axes
plt.xlabel('Error')
plt.ylabel('O-life')

# Show the plot
plt.show()
```
